```python
def groupbyGO():
    genecList=[]
    for k in go2gene.keys():
        genecList.append(len(go2gene[k]))
    genecList = sorted(genecList)
    genel = list(set(genecList))
    outlist = []
    
    for c in genel:
        outlist.append([c,genecList.count(c)]) 
    return outlist
```

```python
ass2count = groupbyGO() # Num genes associated to the number of go terms with that number associated
```
